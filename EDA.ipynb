```python
def impute_age(cols):
  Age = cols[0]
  Pclass = cols[1]

  if pd.isnull(Age):
    if Pclass == 1:
      return 37
    if Pclass == 2:
      return 29
    else:
      return 22
  else:
      return Age
```

Now apply that function!
